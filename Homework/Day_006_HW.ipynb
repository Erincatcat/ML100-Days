# 檢視與處理 Outliers
# 為何會有 outliers, 常見的 outlier 原因

未知值，隨意填補 (約定俗成的代入)，如年齡常見 0、999

可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本


```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#如果不想要每次在jupyter notebook顯示都寫.show()，可先加此
%matplotlib inline 

# 設定 data_path
dir_data = './Data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

# 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
app_train.dtypes.value_counts() #此data frame裡的各欄位資料型態個數
```

```python
# 先篩選數值型的欄位
dtype_select_num = app_train.select_dtypes(include=[np.number])
dtype_select_num.head()
```

```python
# 再把只有 2 值 (通常是 0,1) 的欄位去掉
for cols in dtype_select_num:                        #run loop
    if len(np.unique(dtype_select_num[cols])) == 2:  #若只有2類
        del dtype_select_num[cols]                   #刪除
        
print("Numbers of remain columns" , dtype_select_num.shape[1]) #看剩幾個columns
print("Numbers of remain columns %i" % len(list(dtype_select_num.columns))) #另種方式
```

```python
# 檢視這些欄位的數值範圍
for i in dtype_select_num.columns:
    app_train[[i]].boxplot()
    plt.show()
```

另種表示方式

```python
# 檢視這些欄位的數值範圍
n_rows = int(np.ceil(dtype_select_num.shape[1] / 2))        #以下呈現的圖之rows數,也就是一列兩個圖
fig, axes = plt.subplots(n_rows, 2, figsize=(15, 100)) #subplots分格顯示,製作一個n_rows * 2 的分格、figsize設定圖的大小
for (cols, ax) in zip(dtype_select_num.columns.tolist(), axes.ravel()):
    dtype_select_num.boxplot(column=cols, ax=ax, vert=False)
    ax.set_yticklabels([])
    ax.set_title(cols)
plt.tight_layout()
```

從上面的圖檢查的結果，至少有三個欄位好像有點可疑

AMT_INCOME_TOTAL、REGION_POPULATION_RELATIVE、OBS_60_CNT_SOCIAL_CIRCLE

# Hints: 

Emprical Cumulative Density Plot

[ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0)

[ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())
```

```python
# 繪製 Empirical Cumulative Density Plot (ECDF)

def ECDF(data):   #先定義我的 ECDF
    F=data.value_counts().sort_index().cumsum()
    m=len(data)
    return F / m

cdf = ECDF(app_train['AMT_INCOME_TOTAL'])

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')   
plt.ylabel('ECDF')
#plt.xlim([cdf.index.min(), cdf.index.max() * 1.2]) # 限制顯示圖片的範圍
#plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```

# 補充：Normal dist 的 ECDF


```python
# 最大值落在分布之外
print(app_train['REGION_POPULATION_RELATIVE'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
cdf = ECDF(app_train['REGION_POPULATION_RELATIVE'])

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['REGION_POPULATION_RELATIVE'].hist()
plt.show()

app_train['REGION_POPULATION_RELATIVE'].value_counts()

# 就以這個欄位來說，雖然有資料掉在分布以外，也不算異常，僅代表這間公司在稍微熱鬧的地區有的據點較少，
# 導致 region population relative 在少的部分較為密集，但在大的部分較為疏漏
```

```python
# 最大值落在分布之外
print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)

cdf = ECDF(app_train['OBS_60_CNT_SOCIAL_CIRCLE'])

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min() * 0.95, cdf.index.max() * 1.05])
#plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['OBS_60_CNT_SOCIAL_CIRCLE'].hist()
plt.show()
print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].value_counts().sort_index(ascending = False))
```
