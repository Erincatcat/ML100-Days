# 作業
在精簡深度學習的方式上 : 卷積類神經 (CNN) 採用像素遠近，而遞歸類神經 (RNN) 採用著則是時間遠近  
那麼，既然有著類似的設計精神，兩者是否有可能互相應用呢?

* CNN通常拿來處理影像，類似人類的眼睛功能，辨識不同物體 ex自動駕駛車辨識路況、facebook照片朋友自動標記...等   
* RNN通常拿來處理語言，類似人類的耳朵及嘴巴功能，必須要知道前後文意思串起來代表的意義，解析聲音及語言模式 ex 聊天機器人、siri、google的自動填入功能...等

CNN & RNN 互相應用，舉例：
1. 自動駕駛車除了辨識路況外，也能同時跟車上的人類進行對話
2. 看圖說故事，藉由CNN辨別圖片中有特徵，再利用RNN用一句話說明圖片內容
